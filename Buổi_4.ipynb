```python
#import thu vien
import pandas as pd
#Buoc 1: load data
df = pd.read_excel('Coffee Chain.xlsx')
df.head()
```

```python
#Buoc 2. kiem tra so luong hang cot, kieu du lieu, cac gia tri null
df.info()
```

Các bước kiểm tra dữ liệu
- Kiem tra xem co can giu lai cot Unnamed: 0
- Doi lai kieu du lieu cot Ddate, xu ly o null
- Market Size xu ly null
- Budget Cogs co null
- Budget Sales, Inventory doi lai thanh int/float
- Doi lai ten cot Number of Records(So luong san pham) thanh quantity


```python
#Kiem tra xem co can giu lai cot Unnamed: 0
#cot Unnamed: 0 la cot so thu tu => xoa vi khong can thiet
# df['Unnamed: 0']
del df['Unnamed: 0']
df.info()
```

```python
#doi lai kieu data Ddate->datetime, Budget Sales, Inventory -> float
df['Ddate'] = pd.to_datetime(df['Ddate'])
#doi tu String sang kieu So thi phai replace dau phay truoc, gan regex = true roi moi ep kieu duoc

df['Budget Sales'] = df['Budget Sales'].replace(',' , '', regex = True).astype(float)
df['Inventory'] = df['Inventory'].replace(',' , '', regex = True).astype(int)
df.info()
```

```python
#Xem lai du lieu da duoc thay doi
df.head()
```

- Xử lý dữ liệu null bằng phương pháp thay thế:
  + Sử dụng cột Market_Size làm ví dụ:

```python
df.info()
```

```python
#Kiểm tra cột market size có giá trị null để thay thế không:
df['Market Size'].value_counts()
```

- Như đã thấy ở trên thì số lượng Small Market nhiều hơn Major Market, nếu có 1 số ít ô trống thì fill Small Market vào thì tỉ lệ đúng nó sẽ cao hơn.

- Khi nào sử dụng forward fill(ffill) và backward fill(bfill) để xử lý dữ liệu null:
  + Các dữ liệu liên quan đến Category, Date, ...
  + Chọn cột Ddate để thực hành:

```python
#sd ffill/bfill voi cot date để fill cột null
#ffill sử dụng giá trị của ô liền trước (ở hàng trước)
#bfill thì sử dụng giá trị của ô liền sau (ở hàng sau)
df['Ddate'] =  df['Ddate'].fillna(method = 'ffill', limit = 1)
df.head(6) #limit 6 dữ liệu được xuất ra
```

- Sử dụng mean để fill giá trị bị null:

```python
mean = df['Budget Cogs'].mean()
```

```python
#thay the Budget Cogs = mean
df['Budget Cogs'].replace(np.nan, mean, inplace = True)
df.info()
```

```python
df.head()
```

- Xử lý null bằng cách tạo từ điển:

```python
mk_dict = dict(zip(df['State'], df['Market Size']))
mk_dict
```

- Vì cột State không bị null cột nào nên thay thế Market Size bằng dữ liệu Null

```python
#Tạo Null ở cột Market Size
market_null = df.index[df['Market Size'].isnull()].tolist()
market_null
df.loc[market_null, ]
```

```python
#Thêm dữ liệu từ dictionary vào cột
df['Market Size'] = df['Market Size'].fillna(df['State'].map(mk_dict))
df.info()
```
